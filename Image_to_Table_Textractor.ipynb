```python
import os
from PIL import Image
from textractor import Textractor
from textractor.visualizers.entitylist import EntityList
from textractor.data.constants import TextractFeatures, Direction, DirectionalFinderType
```

```python
image1 = Image.open("/content/Table_1_example.png")
image1
```

```python
image2 = Image.open("/content/Table_2_example.png")
image2
```

```python
extractor = Textractor(profile_name="default")

document1 = extractor.analyze_document(
    file_source=image1,
    features=[TextractFeatures.TABLES],
    save_image=True
)
```

```python
document2 = extractor.analyze_document(
    file_source=image2,
    features=[TextractFeatures.TABLES],
    save_image=True
)
```

```python
document1
```

```python
document2
```

```python
table = EntityList(document1.tables[0])
document1.tables[0].visualize()
```

```python
table2 = EntityList(document2.tables[0])
document2.tables[0].visualize()
```

```python
table[0].to_excel(filepath="AQL_table.xlsx")
```

```python
table2[0].to_excel(filepath="GDP_table.xlsx")
```
